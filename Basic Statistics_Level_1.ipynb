```python
cars=pd.read_csv(r'F:\Excelr Assignments\1_Basic Statistics_Level_1\Cars.csv')
```

```python
cars.head()
```

```python
cars.shape
```

```python
MPG1 = np.round(1 - st.norm.cdf(38, loc= cars.MPG.mean(), scale= cars.MPG.std()),3)
print('P(MPG>38)=',MPG1)
```

```python
MPG2 = np.round(st.norm.cdf(40, loc=cars.MPG.mean() ,scale=cars.MPG.std()),3)
print('P(MPG>40)=',MPG2)
```

```python
MPG3 = np.round(1-st.norm.cdf(20, loc=cars.MPG.mean() ,scale=cars.MPG.std()),3)
print('P(MPG>40)=',MPG3)
```

```python
MPG4= np.round(st.norm.cdf(50, loc=cars.MPG.mean() ,scale=cars.MPG.std()),3)
print('P(MPG>40)=',MPG4)
```

```python
s=MPG4-MPG3
MPG5=s.round(3)
print('P(20<MPG<50)=',MPG5)
```

## Question 22


Calculate the Z scores of 90% confidence interval,94% confidence interval, 60% confidence interval 

```python
print('Z score for 90% Confidence Interval =',np.round(st.norm.ppf(.05),4))
print('Z score for 94% Confidence Interval =',np.round(st.norm.ppf(.03),4))
print('Z score for 60% Confidence Interval =',np.round(st.norm.ppf(.2),4))

```

## Question 23

 Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25

```python
print('T score for 95% Confidence Interval =',np.round(st.t.ppf(0.025,df=24),4))
print('T score for 96% Confidence Interval =',np.round(st.t.ppf(0.04,df=24),4))
print('T score for 99% Confidence Interval =',np.round(st.t.ppf(0.005,df=24),4))

```
